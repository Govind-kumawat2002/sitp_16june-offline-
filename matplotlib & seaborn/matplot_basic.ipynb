```python
x =np.array([1,2,3,4])
label=np.array(["O2","Co2","N","Other"])
myexplode = [0.2,0,0,0]
plt.pie(x , labels=label,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
x =np.array([1,2,3,4])
label=np.array(["O2","Co2","N","Other"])
myexplode = [0.2,0,0,0]
color = ["red", "green", "blue", "yellow"]
plt.pie(x , labels=label,startangle=90,explode=myexplode,shadow=True,colors=color)
plt.show()
```

```python
x =np.array([1,2,3,4])
label=np.array(["O2","Co2","N","Other"])
myexplode = [0.2,0,0,0]
color = ["red", "green", "blue", "yellow"]
plt.pie(x , labels=label,startangle=90,explode=myexplode,shadow=True,colors=color,autopct='%1.2f%%')
plt.show()
```

```python
gases = ['Nitrogen', 'Oxygen', 'Argon', 'Carbon Dioxide', 'Other Gases']
percentages = [78.09, 20.95, 0.93, 0.04, 0.01]
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral', 'gray']
explode = [0, 0.1, 0, 0, 0]

plt.figure(figsize=(8, 8))
plt.pie(
    percentages,
    labels=gases,
    colors=colors,
    autopct='%1.2f%%',
    startangle=90,
    
    explode=explode
) 
plt.show()
```
